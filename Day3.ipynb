```python
import matplotlib.pyplot as plt

#Create data for plotting
x_values = [0,1,2,3,4,5]
y_values = [0,1,4,9,16,25]

#the default graph style for plot is line
plt.plot(x_values,y_values)

#display the graph
plt.show()

plt.bar(x_values,y_values,color="#FFE5B4")
plt.show()

bars = plt.barh([1,2,3,4],[1,2,3,4])
bars[0].set_color("red")
bars[1].set_color("orange")
bars[2].set_color("yellow")
bars[3].set_color("lightgreen")
plt.show()
```

```python
import matplotlib.pyplot as plt
ages=(2,4,6,3,86,34,96,23,35,54,22,11,32,47,62,65,77)
range = (0,100)
bins = 10

plt.hist(ages,bins,range,color="green",histtype="bar",edgecolor="black",alpha=0.55)

plt.xlabel("age")

plt.ylabel("No. of people")

plt.title("My histogram")

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = [0,1,2,3,4,5]
y_values = [0,1,4,9,16,15]

plt.scatter(x_values,y_values,label="stars",s=30,color="skyblue",marker="p")

plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title("My star graph",color="pink")
plt.show()
```

```python
import matplotlib.pyplot as plt
activities=["eat","sleep","work","play"]

slices = [35,80,10,64]

colors=["red","orange","yellow","pink"]

plt.pie(slices,labels=activities,colors=colors,startangle=90,shadow=True,explode=(0,0,1,0),radius=1.2,autopct="%1.lf%%")

plt.title("daily routine")

plt.show()
```
